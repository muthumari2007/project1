```python
import pandas as pd

data = {
    'Department': ['HR', 'IT', 'Sales', 'IT', 'HR', 'Sales', 'IT', 'HR'],
    'Salary': [25000, 40000, 30000, 45000, 28000, 32000, 50000, 27000],
    'Experience': [2, 5, 3, 6, 1, 4, 7, 2],
    'Attrition': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes']
}

df = pd.DataFrame(data)
print(df)
```

```python
df.drop_duplicates(inplace=True)
print("Duplicates Removed")
```

```python
dept = pd.crosstab(df['Department'], df['Attrition'])
print(dept)
```

```python
salary = pd.crosstab(df['Salary'], df['Attrition'])
print(salary)
```

```python
exp = pd.crosstab(df['Experience'], df['Attrition'])
print(exp)
```

```python
import matplotlib.pyplot as plt

dept.plot(kind='bar')
plt.title("Department-wise Attrition")
plt.show()
```

```python
df['Attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.title("Employee Attrition")
plt.show()
```
